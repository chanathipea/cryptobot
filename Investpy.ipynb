```python
def get_funds(my_funds):
    data = ip.funds.get_fund_information(fund=my_funds, country='thailand', as_json=False)
    data = data[['Fund Name','1-Year Change','Category']]
    
    hist_data = ip.funds.get_fund_historical_data(fund=my_funds, 
                                         country='thailand', 
                                         as_json=False, 
                                         from_date = '20/4/2022', 
                                         to_date = '21/4/2022',  
                                         order='ascending',
                                         interval = 'Daily')
    hist_data.reset_index(inplace = True)
    hist_data.rename(columns = {'Date':'Buy Date', 'Close':'Buying Price'}, inplace = True)
    hist = hist_data[['Buy Date', 'Buying Price']].iloc[[0]]
    
    recent_data = ip.funds.get_fund_recent_data(fund=my_funds, 
                                         country='thailand', 
                                         as_json=False,  
                                         order='descending',
                                        interval = 'Daily')
    recent_data.reset_index(inplace = True)
    recent_data.rename(columns = {'Date':'Latest Date', 'Close':'Latest Price'}, inplace = True)
    recent = recent_data[['Latest Date','Latest Price']].iloc[[0]]
    
    fund_name = ip.funds.search_funds(by = 'country', value = 'thailand')
    fund_name = fund_name[fund_name['name'] == my_funds]
    fund_name.reset_index(inplace = True)
    fund_name = fund_name[['symbol']]
    
    df = data.join([hist, recent, fund_name])
    return df
```

```python
def callback():
    # get X-Line-Signature header value
    signature = request.headers['X-Line-Signature']

    # get request body as text
    body = request.get_data(as_text=True)

    # handle webhook body
    try:
        handler.handle(body, signature)
    except InvalidSignatureError:
        print("Invalid signature. Please check your channel access token/channel secret.")
        abort(400)

    return 'OK'
```

```python
def get_value_from_postback(data):
    datas = data.split("&")
    values = []
    for word in datas:
        tmp_value = word.split("=")
        values.append(tmp_value[1])

    return values

if __name__ == '__main__':
    app.run()
```
